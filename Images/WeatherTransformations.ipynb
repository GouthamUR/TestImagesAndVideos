```python
from imagecorruptions import corrupt, get_corruption_names

input_dir = '/content/sample_data'   # Your folder with images
output_dir = '/content/outputs_2'      # Folder to save corrupted images
os.makedirs(output_dir, exist_ok=True)

# Make sure all images in input_dir are processed
image_files = [f for f in os.listdir(input_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

for img_file in image_files:
    img_path = os.path.join(input_dir, img_file)
    image = Image.open(img_path).convert('RGB')       # Ensure RGB
    image_np = np.array(image).astype(np.uint8)       # Convert to np.uint8

    # Loop over all 15 common corruptions
    for corruption in get_corruption_names():
        # Loop over severities 1-5
        for severity in range(1, 6):
            corrupted = corrupt(image_np, corruption_name=corruption, severity=severity)

            # Create folder for each corruption and severity
            save_dir = os.path.join(output_dir, corruption, str(severity))
            os.makedirs(save_dir, exist_ok=True)

            # Save image
            save_path = os.path.join(save_dir, img_file)
            Image.fromarray(corrupted).save(save_path)

print("All corrupted images saved in:", output_dir)
```
